```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
```

```python
file_path_1="/media/yiying/ANDREW_2TB/bee/attacted/20220829-163645.wav"
file_path_2="/media/yiying/ANDREW_2TB/bee/attacted/Yunlin_flu_4_20220829-161654.wav"
file_path_3="/media/yiying/ANDREW_2TB/bee/attacted/Yunlin_flu_4_20220829-163534.wav"
#file_path_3="/media/yiying/Transcend/east/q-2/20210905-172545.wav"
#file_path_4="/media/yiying/Transcend/east/q-4/20210903-142450.wav"
#file_path_5="/media/yiying/Transcend/east/q-4/20210903-172540.wav"
'''file_path_6="/media/yiying/TOSHIBA EXT/Volume/north/q-1/20210823-080435.wav"'''
file = [file_path_1,file_path_2,file_path_3]


for i in file:
    mels = y, sr = librosa.load(i)
    D = np.abs(librosa.stft(y))**2
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)
    
    fig, ax = plt.subplots()
    S_dB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
    fig.colorbar(img, ax=ax, format='%+2.0f dB')
    ax.set(title='queenless-2-before-frequency spectrogram')
    
```
